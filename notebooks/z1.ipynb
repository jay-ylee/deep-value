```python
import plotly.express as px

orange = px.colors.sequential.Oranges
orange.append('rgb(245,130,32)')
```

```python
list(edf.columns.levels[1])
```

```python
import plotly.graph_objects as go

x = ['mean', 'median', 'min', 'max', 'std']

fig = go.Figure()

for i, idx in enumerate(idxs):
    print(0.5 + i / 2 * (len(idxs)-1))
    fig.add_trace(go.Bar(
        y=[edf.loc[idx, ('익월 수익률', 'count')]],
        x=['count'],
        name=idx,
        marker_color=orange[i],
        opacity=0.5 + i / (2 * (len(idxs)-1))
    ))

fig.show()
```
