```python
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
etf = pd.read_csv(
    Path.cwd() / "data" / "historical_prices_etf.csv",
    parse_dates=["_date"],
)
etf.set_index(['_date'], inplace=True)
```

```python
df = etf.stack().reset_index(drop=False)
df.columns = ['_date', '_code', '_value']
df = df.sort_values(['_code', '_date']).reset_index(drop=True)
df['_code'] = df['_code'].str.replace('-US', '')
```

```python
df['_year'] = df['_date'].dt.year
df['_month'] = df['_date'].dt.month
df = df[(df['_date'] >= '2014-01-01') & (df['_date'] <= '2024-04-30')]
```
