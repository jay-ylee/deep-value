```python
from pathlib import Path
import re
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel(
    Path.cwd() / "data" / "알파 이벤트 매매 내역.xlsx",
    sheet_name="Sheet2"
)
```

```python
def remove_brackets(text):
    return re.sub(r'\[.*?\]', '', text)
```

```python
df1 = df1[['종목', '종목명']]
df1['종목명'] = df1['종목명'].apply(remove_brackets).str.strip()
```

```python
df2 = pd.read_excel(
    Path.cwd() / "data" / "알파 이벤트 매매 내역.xlsx",
    sheet_name="Sheet1"
)
```

```python
df1.sort_values('종목').drop_duplicates('종목명', keep='first')
```

```python
pd.merge(df2.iloc[:84], df1.sort_values('종목').drop_duplicates('종목명', keep='first'), on='종목명', how='left').to_excel(Path.cwd() / "data" / "알파 이벤트 매매 내역_2.xlsx")
```
